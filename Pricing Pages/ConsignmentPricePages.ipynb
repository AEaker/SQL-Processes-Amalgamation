```python
import sqlalchemy as sa
import pandas as pd
import datetime as dt
import numpy as np
import os

engine = sa.create_engine("mssql+pymssql://username:password@server/database", echo=True)
```

```python
consigned = {}
i = 1
files = [ f for f in os.listdir('Consignment PP') if os.path.isfile('Consignment PP/'+ f) ]
print(files)
for filename in files:
    consigned[i] = pd.read_excel('Consignment PP/' + filename,sheet_name='Consigned', usecols=['item_id', 'spa_cost', 'rounded_sell', 'desc'])
    i += i

ConsignedConCat = pd.concat(consigned, ignore_index=True)
ConsignedConCat
```

```python
ConsignedConCat.to_sql('aaron_temp_table', con=engine, if_exists='replace')
```
